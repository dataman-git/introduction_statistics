## Discrete random variable

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate tossing a coin five times
num_tosses = 5
num_simulations = 10000
heads_count = np.random.binomial(n=num_tosses, p=0.5, size=num_simulations)

# Plotting the frequency distribution of heads
plt.hist(heads_count, bins=np.arange(-0.5, num_tosses + 1.5, 1), density=True, alpha=0.7, color='blue')
plt.xticks(range(num_tosses + 1))
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Probability Distribution of Heads in 5 Coin Tosses')
plt.grid()
plt.show()
```

## Introduction to Binomial Distribution

### Calculating Probability

```python
from scipy.stats import binom

# Parameters
n = 10  # number of trials
k = 4   # number of successes
p = 0.5 # probability of success

# Calculate probability
probability = binom.pmf(k, n, p)
print(f"P(X = {k}) = {probability:.4f}")
```

### Calculating Cumulative Probability

```python
# Calculate cumulative probability
cumulative_probability = binom.cdf(4, n, p)
print(f"P(X <= {k}) = {cumulative_probability:.4f}")
```

### Variance and Standard Deviation of a Binomial Distribution

```python
import math

# Variance and Standard Deviation
variance = n * p * (1 - p)
std_dev = math.sqrt(variance)

print(f"Variance = {variance:.4f}")
print(f"Standard Deviation = {std_dev:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Range of k values
k_values = np.arange(0, n+1)

# Calculate binomial probabilities for each k
binomial_probs = binom.pmf(k_values, n, p)

# Plotting
plt.bar(k_values, binomial_probs, color='skyblue')
plt.title('Binomial Distribution PMF (n=10, p=0.5)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.show()
```

### The DGP for a binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10          # number of trials
p = 0.5         # probability of success
size = 1000     # number of samples to generate

# Generate samples
samples = np.random.binomial(n, p, size)

# Plotting the results
plt.hist(samples, bins=np.arange(n + 2) - 0.5, density=True, alpha=0.7, color='blue')
plt.title('Histogram of Binomial Distribution Samples')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n + 1))
plt.grid()
plt.show()
```

## The Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0
sigma = 1

# Generating data points
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)

# Plotting the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')
plt.title('Normal Distribution Curve')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Adding vertical lines for mean and standard deviations
plt.axvline(mu, color='red', linestyle='--', label='Mean')
plt.axvline(mu + sigma, color='green', linestyle='--', label='1 Std Dev')
plt.axvline(mu - sigma, color='green', linestyle='--')
plt.axvline(mu + 2*sigma, color='orange', linestyle='--', label='2 Std Dev')
plt.axvline(mu - 2*sigma, color='orange', linestyle='--')
plt.axvline(mu + 3*sigma, color='purple', linestyle='--', label='3 Std Dev')
plt.axvline(mu - 3*sigma, color='purple', linestyle='--')

plt.legend()
plt.grid()
plt.show()
```

### CDF

```python
# Probability of a value falling between -1 and 1
probability = norm.cdf(1, mu, sigma) - norm.cdf(-1, mu, sigma)

print(f"Probability of falling between -1 and 1: {probability:.4f}")
```

### Standard normal distribution

```python
# Parameters for the normal distribution
mu_new = 2
sigma_new = 0.5
X = 1.5

# Calculating the standard score
Z = (X - mu_new) / sigma_new
print(f"Standard score (Z) for X={X}: {Z:.4f}")
```

### The DGP for a normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0         # Mean of the distribution
std_dev = 1      # Standard deviation of the distribution
num_samples = 1000  # Number of samples to generate

# Generating samples from a normal distribution
samples = np.random.normal(mean, std_dev, num_samples)

# Plotting the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Plotting the probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, p, 'k', linewidth=2)

# Adding labels and title
plt.title('Histogram of Samples from a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Display the plot
plt.show()
```

## Poisson distribution

```python
import math

# Parameters
lambda_value = 3  # Average rate (books sold per hour)
k = 5  # Number of occurrences (books sold)

# Poisson probability mass function
poisson_probability = (lambda_value**k * math.exp(-lambda_value)) / math.factorial(k)
print(f"P(X = {k}) = {poisson_probability:.4f}")
```

```python
# Calculate cumulative probability for k = 0 to 5
cumulative_probability = sum((lambda_value**k * math.exp(-lambda_value)) / math.factorial(k) for k in range(6))
print(f"P(X <= 5) = {cumulative_probability:.4f}")
```

```python
from scipy.stats import poisson

# Parameters
lambda_value = 3

# Probability of selling exactly 5 books
probability_exact = poisson.pmf(5, lambda_value)
print(f"P(X = 5) = {probability_exact:.4f}")

# Cumulative probability of selling at most 5 books
cumulative_probability_scipy = poisson.cdf(5, lambda_value)
print(f"P(X <= 5) = {cumulative_probability_scipy:.4f}")
```

```python
# Parameters
lambda_value = 10  # Average calls per hour
k = 8  # Number of calls

# Probability of receiving exactly 8 calls
probability_8_calls = poisson.pmf(k, lambda_value)
print(f"P(X = 8) = {probability_8_calls:.4f}")

# Cumulative probability of receiving at most 8 calls
cumulative_calls = poisson.cdf(k, lambda_value)
print(f"P(X <= 8) = {cumulative_calls:.4f}")
```

### The DGP for Poisson distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set the mean (lambda) of the Poisson distribution
lambda_value = 5

# Generate 1000 random samples from the Poisson distribution
poisson_samples = np.random.poisson(lam=lambda_value, size=1000)

# Display the first 10 samples
print("First 10 Poisson samples:", poisson_samples[:10])

# Plot the histogram of the generated samples
plt.hist(poisson_samples, bins=15, edgecolor='black', alpha=0.7, density=True, label='Sample Data')

# Calculate the theoretical Poisson PMF
x = np.arange(0, max(poisson_samples) + 1)
pmf_values = poisson.pmf(x, mu=lambda_value)

# Plot the theoretical PMF as a line
plt.plot(x, pmf_values, marker='o', color='red', label='Theoretical PMF')

# Adding title and labels
plt.title('Poisson Distribution (lambda = 5)')
plt.xlabel('Number of Events')
plt.ylabel('Density')
plt.legend()
plt.show()
```
