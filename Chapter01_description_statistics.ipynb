### Frequency

#### Code example 1: frequency

```python
# Python code to create a frequency distribution for categorical data
from collections import Counter

# Data: Favorite colors of students
colors = ['Red', 'Blue', 'Blue', 'Green', 'Red', 'Yellow', 'Blue', 
          'Red', 'Green', 'Blue', 'Yellow', 'Red', 'Red', 'Blue', 
          'Green', 'Yellow', 'Blue', 'Red', 'Yellow', 'Red']

# Creating the frequency distribution
freq_distribution = Counter(colors)

# Displaying the frequency distribution
for color, frequency in freq_distribution.items():
    print(f"{color}: {frequency}")
```

#### Code example 2: grouped frequency

```python
import pandas as pd

# Data: Ages of individuals
ages = [18, 22, 19, 21, 24, 23, 20, 22, 19, 21, 23, 22, 24, 
        20, 21, 22, 18, 19, 21, 22, 24, 23, 20, 21, 22, 23, 
        24, 20, 19, 21]

# Creating bins for age groups
bins = [18, 20, 22, 24, 26]
labels = ['18-19', '20-21', '22-23', '24-25']

# Grouping data into intervals
age_groups = pd.cut(ages, bins=bins, labels=labels, right=False)

# Generating frequency distribution
freq_table = age_groups.value_counts().sort_index()

# Displaying the frequency distribution
print(freq_table)
```

#### Code example 3: relative and cumulative frequencies

```python
# Calculating relative and cumulative frequencies
total = sum(freq_table)

# Relative frequency
relative_freq = freq_table / total

# Cumulative frequency
cumulative_freq = freq_table.cumsum()

# Displaying the results
print("Relative Frequency:")
print(relative_freq)
print("Cumulative Frequency:")
print(cumulative_freq)
```

### Bar charts

```python
from collections import Counter
import matplotlib.pyplot as plt

# Data: Favorite fruits of students
fruits = ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 
          'Apple', 'Orange', 'Apple', 'Orange', 'Banana']

# Frequency distribution
fruit_counts = Counter(fruits)

# Plotting the bar chart
plt.bar(fruit_counts.keys(), fruit_counts.values(), color='skyblue')
plt.title('Favorite Fruits of Students')
plt.xlabel('Fruit')
plt.ylabel('Count')
plt.show()
```

### Grouped Bar charts

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Product': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'] * 3,
    'Sales': [150, 200, 250, 300, 220, 270, 350, 400, 180, 230, 290, 310]
}
df = pd.DataFrame(data)   # Create a DataFrame
plt.figure(figsize=(10, 6))  # Set the figure size
sns.barplot(x='Quarter', y='Sales', hue='Product', data=df)
plt.title('Sales Data by Product and Quarter')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.legend(title='Product')
plt.show()
```

### Measures of Center and Spread

#### Measures of Center: Mean

```python
# Data
data = [5, 8, 12, 15, 20]

# Calculating the mean
mean = sum(data) / len(data)
print(f'Mean: {mean}')
```

#### Measures of Center: Median

```python
# Data
data = [5, 8, 12, 15, 20]

# Calculating the median
data.sort()
n = len(data)
median = data[n // 2] if n % 2 == 1 else (data[n // 2 - 1] + data[n // 2]) / 2
print(f'Median: {median}')
```

#### Measures of Center: Mode

```python
from collections import Counter

# Data
data = [1, 2, 2, 2, 3, 4, 4, 4, 5]

# Count frequencies
counts = Counter(data)

# Find the highest frequency
max_freq = max(counts.values())

# Find all values with that frequency
modes = [value for value, freq in counts.items() if freq == max_freq]

print(f'Mode(s): {modes}')
```

#### Measures of Spread: Range

```python
data = [5, 8, 12, 15, 20]
range_value = max(data) - min(data)
print(f'Range: {range_value}')
```

#### Measures of Spread: Variance and standard deviation

```python
import numpy as np
data = [1, 2, 2, 3, 4, 4, 4, 5]
minp = np.min(data)
maxp = np.max(data)
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std = np.std(data)
print('Min', minp)
print('Max', maxp)
print('Mean', mean)
print('Median',median)
print('Variance', variance)
print('Standard deviation', std)
```

### Outliers
#### Code example 1:

```python
import matplotlib.pyplot as plt

# Data: Exam scores
scores = [45, 55, 60, 62, 70, 72, 68, 75, 80, 78, 85, 
          88, 90, 92, 94, 96, 150]

# Plotting the box plot
plt.boxplot(scores)
plt.title('Exam Scores with an Outlier')
plt.ylabel('Scores')

# Adding a scatter point for the outlier
plt.scatter([1], [150], color='red', zorder=5)  # zorder to place the point above the boxplot
plt.text(1, 150, ' Outlier: 150', verticalalignment='bottom', horizontalalignment='right', color='red')

plt.show()
```
