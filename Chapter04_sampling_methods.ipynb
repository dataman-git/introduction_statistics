### Random samples

```python
import random

# Population of 100 students numbered from 1 to 100
population = list(range(1, 101))

# Randomly selecting 10 students
random_sample = random.sample(population, 10)

print("Random Sample of Students:", random_sample)
```

```python
import matplotlib.pyplot as plt
# Visualization
plt.figure(figsize=(10, 6))

# Plot all students in the population
plt.scatter(population, [1]*len(population), color='lightgray', label='Population', s=50)

# Highlight the sampled students
plt.scatter(random_sample, [1]*len(random_sample), color='orange', label='Random Sample', s=100)

# Adding labels and title
plt.xlabel('Student Number')
plt.yticks([])  # Remove y-axis ticks for clarity
plt.title('Random Sampling Visualization')
plt.legend()

# Annotating the sampled points
for sample in random_sample:
    plt.text(sample, 1.02, f'{sample}', ha='center', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()

# Print the sample
print("Random Sample of Students:", random_sample)
```

### Stratified sampling

```python
import numpy as np

# Population divided by class
freshmen = [f'F{i}' for i in range(1, 121)]
sophomores = [f'S{i}' for i in range(1, 51)]
juniors = [f'J{i}' for i in range(1, 31)]

# Combine into a single population
population_stratified = freshmen + sophomores + juniors

# Sample size for each stratum
sample_size = {'Freshmen': 12, 'Sophomores': 5, 'Juniors': 3}

# Selecting samples from each stratum
stratified_sample = []
stratified_sample.extend(random.sample(freshmen, sample_size['Freshmen']))
stratified_sample.extend(random.sample(sophomores, sample_size['Sophomores']))
stratified_sample.extend(random.sample(juniors, sample_size['Juniors']))

print("Stratified Sample of Students:", stratified_sample)
```

#### Stratified sampling visualization

```python
import matplotlib.pyplot as plt
# Data for visualization
population_sizes = [len(freshmen), len(sophomores), len(juniors)]
sample_sizes = [sample_size['Freshmen'], sample_size['Sophomores'], sample_size['Juniors']]
categories = ['Freshmen', 'Sophomores', 'Juniors']

# Plotting the data
x = np.arange(len(categories))  # label locations
width = 0.35  # width of the bars

fig, ax = plt.subplots(figsize=(8, 6))
# Bar for population sizes
bars1 = ax.bar(x - width/2, population_sizes, width, label='Population', color='lightblue')
# Bar for sample sizes
bars2 = ax.bar(x + width/2, sample_sizes, width, label='Sample', color='orange')

# Adding labels, title, and legend
ax.set_xlabel('Student Category')
ax.set_ylabel('Number of Students')
ax.set_title('Stratified Sampling Visualization')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Adding value labels on top of bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 1, f'{int(height)}', ha='center', va='bottom')

# Display the plot
plt.tight_layout()
plt.show()

# Print the sample
print("Stratified Sample of Students:", stratified_sample)
```

### Systematic sampling

```python
# Setting the population
population = list(range(1, 101))

# Randomly choosing a starting point between 1 and 10
start = random.randint(1, 10)

# Selecting every 10th student
systematic_sample = population[start-1::10]

print("Systematic Sample of Students:", systematic_sample)
```

#### Data visualization for systematic sampling

```python
# Visualization setup
plt.figure(figsize=(10, 6))

# Plotting the population
plt.scatter(population, [1]*len(population), color='lightgray', label='Population', s=50)

# Highlighting the systematic sample
plt.scatter(systematic_sample, [1]*len(systematic_sample), color='orange', label='Systematic Sample', s=100)

# Adding labels and title
plt.xlabel('Student ID')
plt.yticks([])  # Remove y-axis ticks for clarity
plt.title('Systematic Sampling Visualization')
plt.legend()

# Annotating the sampled points
for sample in systematic_sample:
    plt.text(sample, 1.02, f'{sample}', ha='center', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()

# Print the sample
print("Systematic Sample of Students:", systematic_sample)
```

### Sampling Distribution of the Sample Mean}

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
mu = 170  # Population mean
sigma = 10  # Population standard deviation
n = 36  # Sample size

# Standard error
se = sigma / np.sqrt(n)

# Generating sampling distribution of the sample mean
x = np.linspace(mu - 4*se, mu + 4*se, 1000)
sampling_distribution = (1 / (se * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / se) ** 2)

# Plotting the sampling distribution
plt.figure(figsize=(10, 6))
plt.plot(x, sampling_distribution, color='blue', label='Sampling Distribution of the Sample Mean')
plt.title('Sampling Distribution of the Sample Mean (n=36)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.axvline(mu, color='red', linestyle='--', label='Population Mean')
plt.legend()
plt.grid()
plt.show()
```

### Sampling Distribution of the Sample Proportion

```python
# Parameters
p = 0.6  # Population proportion
n = 50  # Sample size

# Standard error
se_proportion = np.sqrt((p * (1 - p)) / n)

# Generating sampling distribution of the sample proportion
x_proportion = np.linspace(p - 4*se_proportion, p + 4*se_proportion, 1000)
sampling_distribution_proportion = (1 / (se_proportion * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_proportion - p) / se_proportion) ** 2)

# Plotting the sampling distribution
plt.figure(figsize=(10, 6))
plt.plot(x_proportion, sampling_distribution_proportion, color='green', label='Sampling Distribution of the Sample Proportion')
plt.title('Sampling Distribution of the Sample Proportion (n=50)')
plt.xlabel('Sample Proportion')
plt.ylabel('Density')
plt.axvline(p, color='red', linestyle='--', label='Population Proportion')
plt.legend()
plt.grid()
plt.show()
```

### The Law of Large Numbers

#### Example 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of rolls
num_rolls = 1000

# Simulate rolling a fair die
rolls = np.random.randint(1, 7, num_rolls)
sample_means = np.cumsum(rolls) / np.arange(1, num_rolls + 1)

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(sample_means, label='Sample Mean', color='blue')
plt.axhline(y=3.5, color='red', linestyle='--', label='Expected Value (3.5)')
plt.title('Law of Large Numbers: Sample Mean Convergence')
plt.xlabel('Number of Rolls')
plt.ylabel('Sample Mean')
plt.legend()
plt.grid()
plt.show()
```

#### Example 2

```python
# Number of coin tosses
num_tosses = 1000

# Simulate tossing a fair coin
tosses = np.random.randint(0, 2, num_tosses)
sample_means_coin = np.cumsum(tosses) / np.arange(1, num_tosses + 1)

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(sample_means_coin, label='Sample Mean', color='green')
plt.axhline(y=0.5, color='red', linestyle='--', label='Expected Value (0.5)')
plt.title('Law of Large Numbers: Sample Mean Convergence for Coin Toss')
plt.xlabel('Number of Tosses')
plt.ylabel('Sample Mean')
plt.legend()
plt.grid()
plt.show()
```

### Central Limit Theorem

#### Example 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
population_size = 10000
sample_size = 30
num_samples = 1000

# Generate a uniform population
population = np.random.uniform(low=0, high=100, size=population_size)

# Draw samples and calculate sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]

# Plotting the results
plt.figure(figsize=(12, 6))

# Population distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Sample means distribution
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black')
plt.title('Sampling Distribution of the Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Example 2

```python
# Set different sample sizes
sample_sizes = [5, 10, 30, 50]
num_samples = 1000

plt.figure(figsize=(12, 12))

for i, sample_size in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]
    
    plt.subplot(2, 2, i + 1)
    plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
    plt.title(f'Sampling Distribution of the Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Example 3

```python
# Heights data
heights = np.array([160, 165, 170, 175, 180, 155, 168, 172, 169, 178,163, 177, 182, 174, 171, 166, 167, 173, 164, 179,162, 161, 175, 176, 168, 170, 169, 171, 174, 177])

# Calculate sample mean and standard error
sample_mean = np.mean(heights)
sample_std = np.std(heights, ddof=1)
sample_size = len(heights)
standard_error = sample_std / np.sqrt(sample_size)

print(f"Sample Mean: {sample_mean:.2f} cm")
print(f"Standard Error: {standard_error:.2f} cm")
```

```python
# Visualization
plt.figure(figsize=(10, 6))

# Histogram for the heights
plt.hist(heights, bins=10, color='skyblue', edgecolor='black', alpha=0.7, label='Height Distribution')

# Adding sample mean line
plt.axvline(sample_mean, color='orange', linestyle='--', linewidth=2, label=f'Sample Mean: {sample_mean:.2f} cm')

# Adding labels for standard error
plt.text(sample_mean + standard_error, 2, f'+SE', color='red', fontsize=10, ha='center')
plt.text(sample_mean - standard_error, 2, f'-SE', color='red', fontsize=10, ha='center')

# Highlight standard error range
plt.fill_betweenx([0, 5], sample_mean - standard_error, sample_mean + standard_error, color='red', alpha=0.2, label='Standard Error')

# Plot formatting
plt.title('Height Distribution with Sample Mean and Standard Error', fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Applications

#### Application 1: Average Delivery Time of Packages

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 500 random delivery times between 6 and 10 hours
delivery_times = np.random.uniform(6, 10, 500)

# Take 30 random samples, each of size 50, and calculate the means
sample_means = [np.mean(np.random.choice(delivery_times, 50)) for _ in range(30)]

# Calculate the mean of the sample means
mean_of_sample_means = np.mean(sample_means)

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the original data on the left
axes[0].hist(delivery_times, bins=20, edgecolor='black', color='orange')
axes[0].set_xlabel('Delivery Time (Hours)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Original Data Distribution')

# Plot the distribution of sample means on the right
axes[1].hist(sample_means, bins=10, edgecolor='black', color='lightpink')
axes[1].axvline(mean_of_sample_means, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_of_sample_means:.2f}')
axes[1].set_xlabel('Sample Mean Delivery Time (Hours)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Distribution of Sample Means')
axes[1].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

#### Application 2: Daily Steps Taken by Students

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1,000 step counts following an exponential distribution
step_counts = np.random.exponential(scale=7000, size=1000)

# Take 50 random samples, each of size 30, and calculate the means
sample_means = [np.mean(np.random.choice(step_counts, 30)) for _ in range(50)]

# Calculate the mean of the sample means
mean_of_sample_means = np.mean(sample_means)

# Create subplots for side-by-side graphs
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left graph: Original data histogram
axes[0].hist(step_counts, bins=20, edgecolor='black', color='skyblue', alpha=0.7)
axes[0].set_xlabel('Step Counts', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('Original Step Count Distribution', fontsize=14)

# Right graph: Distribution of sample means
axes[1].hist(sample_means, bins=10, edgecolor='black', color='lightgreen', alpha=0.7)
axes[1].axvline(mean_of_sample_means, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_of_sample_means:.2f}')
axes[1].set_xlabel('Sample Mean Steps', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title('Distribution of Sample Means', fontsize=14)
axes[1].legend()

# Adjust layout and display
plt.tight_layout()
plt.show()
```

#### Application 3: Average Score on Quiz Retakes

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt

# Generate 300 random quiz scores following a beta distribution
quiz_scores = beta.rvs(2, 5, size=300) * 100

# Take 40 random samples, each of size 25, and calculate the means
sample_means = [np.mean(np.random.choice(quiz_scores, 25)) for _ in range(40)]

# Calculate the mean of the sample means
mean_of_sample_means = np.mean(sample_means)

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Left plot: Original data distribution
axes[0].hist(quiz_scores, bins=15, edgecolor='black', color='skyblue', alpha=0.7)
axes[0].set_title('Original Quiz Scores Distribution')
axes[0].set_xlabel('Quiz Scores')
axes[0].set_ylabel('Frequency')

# Right plot: Distribution of sample means
axes[1].hist(sample_means, bins=10, edgecolor='black', color='orange', alpha=0.7)
axes[1].axvline(mean_of_sample_means, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_of_sample_means:.2f}')
axes[1].set_title('Distribution of Sample Means')
axes[1].set_xlabel('Sample Mean Quiz Scores')
axes[1].set_ylabel('Frequency')
axes[1].legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
